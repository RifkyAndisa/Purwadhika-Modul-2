```
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import requests

url = 'http://jsonplaceholder.typicode.com/users'
user = requests.get(url)
user
```

```
user_json = user.json()
```

```
user_json
```

```
df = pd.DataFrame(user_json)
df.head()
```

```
df.address = list(map(lambda x: x['street']+", "+x['suite']+", "+x['city']+", "+x['zipcode'], df.address.values))
df.company = list(map(lambda x: x['name'], df.company.values))
df.head()
```
